### Question 1 (Easy): Count Vowels and Their Occurrences

#### Problem Statement:Write a function that takes a string and returns a dictionary with vowels as keys and their counts as values. Ignore case sensitivity.

```python
# Example:
# Input: "Hello World"
# Output: {'e': 1, 'o': 2, 'a': 0, 'i': 0, 'u': 0}
# Or just non-zero counts: {'e': 1, 'o': 2}

def count_vowels(text):
    vowels = 'aeiou'
    text = text.lower()
    result = {} 
    for vowel in vowels:
        count = text.count(vowel)
        if count>0:
            result[vowel]=count
    return result

## what if the chars to be return is also a list 
def count_chars(text,chars):
    chars = chars.lower()
    text= text.lower() 
    result = {}
    for char in chars:
        count = text.count(char)
        if count>0:
            result[char]= count
    return result
```

```python
print(count_vowels('Hello World!'))
print(count_chars('Hello World','oh!'))
```
